```python
import pdfplumber
import pandas as pd
import re
import os
from google.colab import files

# Upload all 5 PDFs
uploaded = files.upload()

def clean_text(s):
    if s is None:
        return None
    return re.sub(r"\s+", " ", str(s)).strip()

def normalize_date(value):
    if not value:
        return None
    value = clean_text(value)
    dt = pd.to_datetime(value, errors="coerce")
    if pd.isna(dt):
        return value
    return dt.strftime("%Y-%m-%d")

def parse_county_from_filename(filename):
    base = os.path.splitext(os.path.basename(filename))[0]
    m = re.search(r"in\s+(.+)$", base, re.IGNORECASE)
    return m.group(1).strip() if m else base

def extract_pdf_lines(pdf_path):
    lines = []
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            txt = page.extract_text()
            if txt:
                page_lines = [clean_text(x) for x in txt.splitlines()]
                lines.extend(page_lines)
    return lines

def is_noise_line(line):
    if not line:
        return True

    patterns = [
        r"^OSTDS Commercial and IM Operating Permit List",
        r"County Health Department",
        r"^Run Date:",
        r"^DISCLAIMER:",
        r"^Page \d+",
        r"^Permit Number Address Zip",
        r"^Permit$",
        r"^Permit Exp\.$",
        r"^Permit Issue$",
        r"^Insp\.$",
        r"^City Freq\.$",
        r"^Last Insp\.$",
        r"^Name of System$",
        r"^Svc Agmt$",
        r"^Svc Agmt ME Name Exp Date$",
        r"^Aerobic$",
        r"^Performance Based$",
        r"^Conventional$",
        r"^Other$",
        r"^Totals:",
        r"^Grand Totals",
    ]
    return any(re.search(p, line, re.IGNORECASE) for p in patterns)

def looks_like_start_line(line):
    # Handles:
    # 36-QA-28908
    # 36-QB-29285
    # 36-QD-3084526
    # 36-QN-1001389
    # 43-QS-00110
    # 47-Q1-2848615
    # 47-QX-1496530
    # 50-QW-00033
    # 50-QF-973927
    # 56-QF-00136
    return bool(re.match(r"^\d{2}-[A-Z0-9]{2}-\d+\b", clean_text(line), re.IGNORECASE))

def split_start_line_tokens(line):
    """
    Parse the first row of each two-line record.

    General shape:
    PERMIT_NUMBER <name/address/city> ZIP PERMIT_EXP PERMIT_ISSUE [FREQ] [LAST_INSP]

    Some rows are missing FREQ and/or LAST_INSP.
    Some have 01/01/01 style dates.
    """
    line = clean_text(line)
    tokens = line.split()

    permit_number = tokens[0]
    rest = tokens[1:]

    # Find ZIP from the right by looking for first 5-digit token
    zip_idx = None
    for i in range(len(rest) - 1, -1, -1):
        if re.fullmatch(r"\d{5}", rest[i]):
            zip_idx = i
            break

    if zip_idx is None:
        return None

    body_tokens = rest[:zip_idx]
    zip_code = rest[zip_idx]
    tail = rest[zip_idx + 1:]

    # collect dates from tail
    date_positions = []
    for i, tok in enumerate(tail):
        if re.fullmatch(r"\d{1,2}/\d{1,2}/\d{2,4}", tok):
            date_positions.append(i)

    permit_exp = tail[date_positions[0]] if len(date_positions) >= 1 else None
    permit_issue = tail[date_positions[1]] if len(date_positions) >= 2 else None
    last_insp = tail[date_positions[2]] if len(date_positions) >= 3 else None

    # inspection frequency is usually the first non-date token after permit_issue
    inspection_frequency = None
    if len(date_positions) >= 2:
        after_issue_idx = date_positions[1] + 1
        if after_issue_idx < len(tail):
            candidate = tail[after_issue_idx]
            if not re.fullmatch(r"\d{1,2}/\d{1,2}/\d{2,4}", candidate):
                inspection_frequency = candidate

    body = " ".join(body_tokens)

    # split body into name + address at first street number
    num_match = re.search(r"\b\d+\b", body)
    if num_match:
        name = body[:num_match.start()].strip(" ,")
        address = body[num_match.start():].strip()
    else:
        name = body.strip(" ,")
        address = None

    return {
        "permit_number": permit_number,
        "name": clean_text(name),
        "address": clean_text(address + " " + zip_code if address else zip_code),
        "last_inspection": normalize_date(last_insp),
        "permit_expiration": normalize_date(permit_exp),
        "permit_issue": normalize_date(permit_issue),
        "inspection_frequency": clean_text(inspection_frequency),
    }

def parse_detail_line(line):
    """
    Parse second row:
    <name_of_system> <maintenance_entity> <license> [Yes|No] [exp_date]

    But some rows are missing:
    - svc_agmt yes/no
    - exp_date
    - even most of the detail row
    """
    line = clean_text(line)

    if not line:
        return {
            "name_of_system": None,
            "svc_agmt": None,
            "service_agreement_expiration_date": None,
        }

    # date at end
    exp_date = None
    m = re.search(r"(\d{1,2}/\d{1,2}/\d{2,4})$", line)
    if m:
        exp_date = normalize_date(m.group(1))
        line = clean_text(line[:m.start()])

    # Yes/No at end
    svc_agmt = None
    m = re.search(r"\b(Yes|No)\b$", line, re.IGNORECASE)
    if m:
        svc_agmt = m.group(1).title()
        line = clean_text(line[:m.start()])

    # remove trailing license number
    m = re.search(r"\b\d{4,}\b$", line)
    if m:
        line = clean_text(line[:m.start()])

    name_of_system = line if line else None

    return {
        "name_of_system": clean_text(name_of_system),
        "svc_agmt": svc_agmt,
        "service_agreement_expiration_date": exp_date,
    }

def parse_pdf(pdf_path):
    raw_lines = extract_pdf_lines(pdf_path)
    lines = [x for x in raw_lines if not is_noise_line(x)]

    rows = []
    skipped_start_lines = []

    i = 0
    while i < len(lines):
        line = lines[i]

        if looks_like_start_line(line):
            start = split_start_line_tokens(line)

            if start is None:
                skipped_start_lines.append(line)
                i += 1
                continue

            detail = {
                "name_of_system": None,
                "svc_agmt": None,
                "service_agreement_expiration_date": None,
            }

            # If next line is not another start line, treat as detail line
            if i + 1 < len(lines) and not looks_like_start_line(lines[i + 1]):
                detail = parse_detail_line(lines[i + 1])
                i += 1

            rows.append({
                "county": parse_county_from_filename(pdf_path),
                "source_file": os.path.basename(pdf_path),
                "permit_number": start["permit_number"],
                "name": start["name"],
                "name_of_system": detail["name_of_system"],
                "address": start["address"],
                "last_inspection": start["last_inspection"],
                "service_agreement_expiration_date": detail["service_agreement_expiration_date"],
                "svc_agmt": detail["svc_agmt"],
                "permit_expiration": start["permit_expiration"],
                "permit_issue": start["permit_issue"],
                "inspection_frequency": start["inspection_frequency"],
            })

        i += 1

    return rows, skipped_start_lines

all_rows = []
summary_rows = []
skipped_rows = []

for filename in uploaded.keys():
    try:
        rows, skipped = parse_pdf(filename)
        all_rows.extend(rows)

        summary_rows.append({
            "source_file": os.path.basename(filename),
            "county": parse_county_from_filename(filename),
            "rows_parsed": len(rows),
            "skipped_start_lines": len(skipped),
        })

        for s in skipped:
            skipped_rows.append({
                "source_file": os.path.basename(filename),
                "county": parse_county_from_filename(filename),
                "skipped_start_line": s,
            })

        print(f"{filename}: parsed {len(rows)} rows, skipped {len(skipped)}")

    except Exception as e:
        print(f"{filename}: ERROR - {e}")
        summary_rows.append({
            "source_file": os.path.basename(filename),
            "county": parse_county_from_filename(filename),
            "rows_parsed": 0,
            "skipped_start_lines": None,
            "error": str(e),
        })

df = pd.DataFrame(all_rows)
summary_df = pd.DataFrame(summary_rows)
skipped_df = pd.DataFrame(skipped_rows)

final_columns = [
    "name",
    "name_of_system",
    "address",
    "last_inspection",
    "service_agreement_expiration_date",
    "svc_agmt",
    "county",
    "permit_number",
    "source_file",
]

if not df.empty:
    df = df[[c for c in final_columns if c in df.columns]]

output_file = "permit_rows1.xlsx"

with pd.ExcelWriter(output_file, engine="openpyxl") as writer:
    df.to_excel(writer, index=False, sheet_name="all_rows")
    summary_df.to_excel(writer, index=False, sheet_name="summary")
    skipped_df.to_excel(writer, index=False, sheet_name="skipped_rows")

print("\nTOTAL ROWS PARSED:", len(df))
print("\nSUMMARY:")
print(summary_df)

files.download(output_file)
```
